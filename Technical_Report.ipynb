## Introduction

This project uses the Water Pollution and Disease dataset from Kaggle to analyze global water quality patterns and classify the safety of water samples. The dataset includes chemical, biological, physical, and geographic indicators—such as nitrate levels, lead concentration, bacteria count, pH, dissolved oxygen, region, and water source. Using simplified WHO and EPA drinking-water guidelines, we derived two labels for supervised learning: Risk Level (Low, Medium, High) and Contaminated? (Yes/No). These labels represent whether a water sample exceeds critical public-health thresholds.

Our primary classification task is to predict the Risk Level of a water sample using measurable environmental attributes. We implemented and compared three classifiers studied in CPSC 322: a Decision Tree, a K-Nearest Neighbors (KNN) classifier, and a custom-built Random Forest classifier implemented using test-driven development. Based on our experimental results, the Random Forest approach performed the best overall, achieving the highest accuracy and producing more stable predictions across multiple runs. These findings demonstrate that ensemble methods are more effective than single-tree or distance-based models for this type of environmental classification task.

## Data Analysis

The dataset contains 3,000 instances, each representing a water-quality measurement from a specific country, region, and water source. It includes a mix of numeric attributes—such as contaminant level (ppm), nitrate level (mg/L), lead concentration (µg/L), bacteria count (CFU/mL), dissolved oxygen (mg/L), and pH—and categorical attributes including Country, Region, Water Source Type, and Treatment Method. The derived label Risk Level (Low, Medium, High) is used as our primary target in classification.

Summary statistics (Figure 1) show that several chemical indicators exhibit substantial variance across samples, particularly bacteria count, which is heavily right-skewed. Nitrate and lead concentrations also show strong variation, with some samples exceeding common safety guidelines by significant margins. The distribution of Risk Levels (Figure 2) reveals that most samples fall into the Medium or High categories, while the Low-risk class represents a smaller portion of the dataset.

To better understand relationships between features and labels, we produced several visualizations. A boxplot of nitrate level grouped by Risk Level (Figure 3) shows a clear upward trend, with High-risk samples exhibiting substantially elevated nitrate levels compared to Low-risk samples. A similar pattern appears in bacteria count and lead concentration. The correlation heatmap (Figure 4) highlights strong relationships between various contamination measures and our derived Water Quality Violations metric. These visual patterns confirm that the attributes used to derive the labels are meaningful predictors and justify their inclusion in our classification models.

## Classification Results

To classify Risk Level, we implemented three machine-learning approaches: a Decision Tree, a K-Nearest Neighbors classifier (k=5), and a custom Random Forest classifier (with parameters N, M, and F tuned over multiple runs). Before training, all categorical attributes were one-hot encoded, and a stratified train/test split preserved label proportions across sets. We evaluated each classifier using accuracy and confusion matrices, and for the Random Forest, we ran the model repeatedly to account for randomness in the bootstrap and feature-sampling process.

The Decision Tree produced reasonable accuracy but showed signs of overfitting, often misclassifying borderline Medium-risk samples as High. The KNN classifier performed slightly worse, likely due to high dimensionality from one-hot encoding and overlapping distances between Medium and High classes. In contrast, our Random Forest classifier consistently achieved the highest accuracy, benefiting from bootstrapping and attribute subsampling, which reduce variance and improve generalization. Its confusion matrix showed fewer misclassifications, especially for High-risk samples, and produced more stable predictions across runs.

Overall, the Random Forest clearly outperformed the other two classifiers and is the recommended model for this task. These results align with expectations from ensemble-learning theory: aggregating multiple diverse trees yields a more robust decision boundary than any individual model.

## Conclusion

This project examined water quality using a real-world dataset and applied machine-learning techniques to classify contamination risk. The dataset presented several challenges, including class imbalance between Low and High risk, skewed numerical distributions (especially bacteria count), and high dimensionality caused by categorical one-hot encoding. Nonetheless, the classifiers performed reasonably well, with the Random Forest achieving the best overall accuracy and reliability.

The main strengths of this approach include the interpretability of the derived thresholds, the meaningful correlation between features and labels, and the robustness of ensemble methods. However, improvements could be made. Feature scaling may help KNN performance, PCA or embedding techniques could reduce dimensionality, and more sophisticated models—such as Gradient Boosting, XGBoost, or SVMs—could improve classification further. Additional temporal or environmental variables may also increase predictive power.

Overall, this work demonstrates that machine learning can effectively support environmental monitoring and may be valuable for identifying unsafe water conditions in real-world public-health applications.
